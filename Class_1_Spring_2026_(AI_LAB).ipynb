### 2. Theoretical Questions & Answers

####Q1: What is data preprocessing, and why is it important in AI?


Data preprocessing is the technique of transforming raw data into a clean and structured format before training an AI model. It is essential because:



*   AI models perform poorly on unclean data.
*   Missing or incorrect values can lead to biased results.

*   It ensures consistency, improves accuracy, and reduces computation time.







####Q2: What are the main steps in data preprocessing?

The main steps in data preprocessing are:

* Data Cleaning – Handling missing values, removing duplicates, correcting incorrect data.
* Data Transformation – Encoding categorical variables, normalizing numerical features.
* Feature Selection – Choosing relevant features for model training.
* Feature Scaling – Standardizing numerical data to improve model performance.

####Q3: What are the different methods to handle missing values?

* Remove missing values – Drop rows or columns with missing data.
* Replace missing values – Use mean, median, or mode to fill in missing values.
* Use interpolation – Estimate missing values based on other data points.

####Q4: What is the difference between categorical and numerical data?

* Categorical Data – Non-numeric data that represents categories (e.g., Gender, City).
* Numerical Data – Data that consists of numbers and can be measured (e.g., Age, Salary).

####Q5: What are the methods of encoding categorical variables?

* One-Hot Encoding: Converts categorical values into binary columns
(e.g., Male → [1,0], Female → [0,1]).
* Label Encoding: Assigns a numeric value to each category (e.g., Male → 0, Female → 1).

####Q6: What is feature scaling, and why is it necessary?

Feature scaling standardizes data so that all numerical values have the same scale. It is necessary because:

* Prevents large values from dominating small ones.
* Improves model convergence and accuracy.
* Common methods include Min-Max Scaling and Standardization.


####Q7: What is the difference between normalization and standardization?


* Normalization (Min-Max Scaling): Rescales data between [0,1] using (X - min) / (max - min).
* Standardization (Z-score Scaling): Centers data around 0 with a standard deviation of 1 using (X - mean) / std.

###3.1. Working with NumPy
Exercise 1: Creating NumPy Arrays

📌 Task: Create a NumPy array and perform operations on it.

```python
import numpy as np

arr = np.arange(1, 13)  # Create an array from 1 to 12
print("Original Array:\n", arr)

reshaped_arr = arr.reshape(3, 4)  # Reshape into a 3x4 matrix
print("\nReshaped (3x4) Array:\n", reshaped_arr)
```

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

print("Matrix A:\n", A)
print("\nMatrix B:\n", B)

# Element-wise multiplication
element_wise = A * B
print("\nElement-wise Multiplication:\n", element_wise)

# Matrix multiplication (dot product)
matrix_mult = np.dot(A, B)
print("\nMatrix Multiplication:\n", matrix_mult)

# Transpose of a matrix
transpose_A = np.transpose(A)
print("\nTranspose of A:\n", transpose_A)
```

```python
data = np.array([10, 20, 30, 40, 50, 60, 70, 80, 90])

print("Mean:", np.mean(data))
print("Median:", np.median(data))
print("Standard Deviation:", np.std(data))
print("Variance:", np.var(data))
```

###3.2 Working with Pandas

📌 Task: Create a Pandas DataFrame and perform multiple operations.

```python
import pandas as pd

# Creating a DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Emma'],
        'Age': [25, 30, 35, 40, 45],
        'Salary': [50000, 60000, 70000, 80000, None]}

df = pd.DataFrame(data)
print("Original DataFrame:\n", df)

# Filtering Data
filtered_df = df[df['Age'] > 30]
print("\nFiltered Data (Age > 30):\n", filtered_df)

# Adding a new column
df['Bonus'] = df['Salary'] * 0.1
print("\nDataFrame with Bonus Column:\n", df)
```

###6. Handling Missing Values
The Iris dataset does not have missing values, but if there were, we could:

* Remove rows with missing values (df.dropna()).
* Fill missing values with the mean (df.fillna(df.mean())).

Exercise 6: Simulating Missing Data

📌 Task: Introduce missing values in sepal_width and handle them.

```python
import numpy as np

# Introduce missing values in 'sepal_width' (randomly select 5 rows)
df.loc[np.random.choice(df.index, 5), "sepal_width"] = np.nan

# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Fill missing values with column mean
df["sepal_width"].fillna(df["sepal_width"].mean(), inplace=True)

# Verify missing values are handled
print("\nMissing Values after Handling:")
print(df.isnull().sum())
```

###7. Handling Duplicates

Exercise 7: Removing Duplicate Entries

📌 Task: Identify and remove duplicate rows (if any).

```python
# Check for duplicate rows
print("\nNumber of Duplicates:", df.duplicated().sum())

duplicates = df[df.duplicated()] #This excludes the first occurrence of each duplicate row.

print(duplicates)

duplicates_all = df[df.duplicated(keep=False)] #This includes the first occurrence of each duplicate row.
print("All Duplicate Rows:\n", duplicates_all)

# Remove duplicates
df = df.drop_duplicates()

print("\nDataset shape after Removing Duplicates:", df.shape)
```

###8. Encoding Categorical Variables
Exercise 8: Convert 'species' into Numeric Data

📌 Task: Convert categorical values
(species) into numerical values.

🔹 Solution: Use Label Encoding since species is an ordinal category.

```python
from sklearn.preprocessing import LabelEncoder

# Initialize encoder
encoder = LabelEncoder()

# Convert 'species' column to numeric labels
df["species"] = encoder.fit_transform(df["species"])

# Display first few rows
print("\nDataset after Encoding:\n", df.head())
```

###9. Feature Scaling
Exercise 9: Normalizing Numerical Features

📌 Task: Normalize sepal_length and petal_length between 0 and 1.

🔹 Solution: Use Min-Max Scaling.



```python
from sklearn.preprocessing import MinMaxScaler

# Initialize scaler
scaler = MinMaxScaler()

# Select numerical columns to scale
columns_to_scale = ["sepal_length", "petal_length"]
df_scaled = df.copy()
df_scaled[columns_to_scale] = scaler.fit_transform(df[columns_to_scale])

# Display first few rows
print("\nDataset after Scaling:\n", df_scaled.head())
```

###10. Data Filtering
Exercise 10: Selecting Flowers Based on Petal Length

📌 Task: Filter flowers with a petal length greater than 4 cm.

```python
# Filter flowers with petal length > 4 cm
filtered_df = df[df["petal_length"] > 4]  # Since it's scaled

# Display first few rows
print("\nFiltered Dataset:\n", filtered_df.head())
```
